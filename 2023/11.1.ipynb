```python
import numpy as np
```

```python
with open('input11.txt') as f:
    data = f.read()
mat = data.splitlines()
mat = list(map(list, mat))
mat = np.matrix(mat)
print(mat)
```

```python
n, m = mat.shape
ax_0 = []
ax_1 = []

for i in range(n):
    if "#" not in mat[i,:]:
        ax_0.append(i)

for j in range(m):
    if "#" not in mat[:,j]:
        ax_1.append(j)

for i in range(len(ax_0)-1, -1, -1):
    mat = np.insert(mat, ax_0[i], ".", axis=0)

for i in range(len(ax_1)-1, -1, -1):
    mat = np.insert(mat, ax_1[i], ".", axis=1)
```

```python
galaxies = []

n, m = mat.shape

for i in range(n):
    for j in range(m):
        if mat[i,j] == "#":
            galaxies.append((i,j))
```

```python
galaxy_num = len(galaxies)
distances = np.zeros((galaxy_num, galaxy_num))

for i, item1 in enumerate(galaxies):
    for j, item2 in enumerate(galaxies):
        distances[i,j] = abs(galaxies[i][0] - galaxies[j][0]) + abs(galaxies[i][1] - galaxies[j][1])
```

```python
sum(sum(distances))/2
```
