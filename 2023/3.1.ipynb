```python
num_dict = {m.start(0):int(m.group(0)) for m in re.finditer("\d+", data)}
```

```python
def next_to_special_char(data, x, y):
    special = "[$&+,:;=?@#|'<>-^*()%!/#$%&*+-/=@]"
    n, m = data.shape
    n = n - 1
    m = m - 1

    if x == n and y == m:
        if data[x-1, y-1] in special or data[x-1, y] in special or data[x, y-1] in special:
            return(True)

    elif x == n:
        if data[x-1, y] in special or data[x-1, y-1] in special or data[x-1, y+1] in special or data[x, y-1] in special or data[x, y+1] in special:
            return(True)

    elif y == m:
        if data[x-1, y] in special or data[x-1, y-1] in special or data[x, y-1] in special or data[x+1, y-1] in special or data[x+1, y] in special:
            return(True)
        
    elif x == 0 and y == 0:
        if data[x, y+1] in special or data[x+1, y+1] in special or data[x+1, y] in special:
            return(True)
    
    elif x == 0:
        if data[x, y-1] in special or data[x+1, y-1] in special or data[x+1, y] in special or data[x+1, y+1] in special or data[x, y+1] in special:
            return(True)
    
    elif y == 0:
        if data[x-1, y] in special or data[x-1, y+1] in special or data[x, y+1] in special or data[x+1, y+1] in special or data[x+1, y] in special:
            return(True)
    
    else:
        if data[x-1, y] in special or data[x-1, y+1] in special or data[x, y+1] in special or data[x+1, y+1] in special or data[x+1, y] in special or data[x+1, y-1] in special or data[x, y-1] in special or data[x-1, y-1] in special:
            return(True)
    return(False)
```

```python
n = len(num_dict)
dim = mat.shape[1]
check = []

for i in num_dict:
    num_len = len(str(num_dict[i]))
    check_temp = []
    
    for j in range(num_len):
        x, y = divmod(i + j, dim)

        if next_to_special_char(mat, x, y):
            check_temp.append(True)
        else:
            check_temp.append(False)
    check.append(check_temp)

check
```
