```python
import pandas as pd
from itertools import chain
```

```python
with open('input7.txt') as f:
    data = f.read().splitlines()
data = [sec.split() for sec in data]

for i in range(len(data)):
    data[i][0] = [char for char in data[i][0]]
    data[i][1] = int(data[i][1])
data
```

```python
card_to_num = {"T": 10, "J": 1, "Q": 12, "K": 13, "A": 14}
```

```python
n = len(data)

for i in range(n):
    m = len(data[i][0])

    for j in range(m):
        if data[i][0][j].isdigit():
            data[i][0][j] = int(data[i][0][j])
        else:
            data[i][0][j] = card_to_num[data[i][0][j]]
        
    data[i][1] = [data[i][1]]
    data[i] = list(chain(*data[i]))

data = pd.DataFrame(data)
data
```

```python
def hand_type(df):
    n = len(df)
    df[6] = 0

    for i in range(n):
        my_dict = {j:list(data.iloc[i][0:5]).count(j) for j in list(data.iloc[i])[0:5]}
        temp = []
        jokers = 0


        if 1 in my_dict:
            jokers = my_dict[1]
            del my_dict[1]

        if len(my_dict) != 0:
            for j in range(len(my_dict)):
                temp.append(my_dict[list(my_dict)[j]])
            
            print(temp, jokers)
            index = temp.index(max(temp))
            temp[index] = temp[index] + jokers
        else:
            temp = [5]
        

        if 5 in temp:
            score = 7

        elif 4 in temp:
            score = 6
        
        elif 3 in temp and 2 in temp:
            score = 5
        
        elif 3 in temp:
            score = 4
        
        elif temp.count(2) == 2:
            score = 3
        
        elif max(temp) == 2:
            score = 2
        
        elif max(temp) == 1:
            score = 1


        df[6][i] = score

    return(df)
```

```python
data = hand_type(data)
```
