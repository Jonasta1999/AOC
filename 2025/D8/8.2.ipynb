```python
import numpy as np
import math
```

```python
with open('input8.txt') as f:
    data = f.read().splitlines()
    data = [list(map(int, x.split(","))) for x in data]
n = len(data)
distances = np.zeros((n, n))
```

```python
def find_euclidean_distance(coord1, coord2):
    x1, y1, z1 = coord1
    x2, y2, z2 = coord2

    distance = math.sqrt((x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)
    return distance
```

```python
def find_all_distances(matrix):
    global data
    for i in range(n):
        for j in range(i, n):
            if i == j:
                matrix[i,j] = float("inf")
                continue
            coord1 = data[i]
            coord2 = data[j]

            distance = find_euclidean_distance(coord1, coord2)

            matrix[i,j] = distance
            matrix[j,i] = distance
    return matrix
```

```python
def find_shortest_distance_and_remove(matrix):
    min_index = np.argmin(np.array(matrix))
    n_min, m_min = np.unravel_index(min_index, matrix.shape)
    matrix[n_min, m_min] = float("inf")
    matrix[m_min, n_min] = float("inf")
    return n_min, m_min, matrix
```

```python
def insert_cicuits(x, y, circuits):
    if not circuits:
        return [[x,y]]
    circuits.append([x,y])
    index = len(circuits)-1
    to_remove = [index]
    new_circuit = [x, y]
    for i, circuit in enumerate(circuits):
        if i == index:
            continue
        if x in circuit or y in circuit:
            to_remove.append(i)
            new_circuit += circuit
    new_circuit = list(set(new_circuit))
    for index in sorted(to_remove, reverse=True):
        del circuits[index]
    circuits.append(new_circuit)
    return circuits       
```

```python
distances = find_all_distances(distances)
iter = 1000
data_len = len(data)
res = []
first_run = True
while first_run or not (len(res[0]) == data_len):
    x, y, distances = find_shortest_distance_and_remove(distances)
    x, y = int(x), int(y)
    res = insert_cicuits(x, y, res)
    first_run = False
```
