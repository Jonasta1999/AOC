```python
import numpy as np
```

```python
with open('input6.txt') as f:
    data = f.read().splitlines()
    nums = data[:-1]
    operators = data[-1]
    operators = list(filter(None, operators.split(" ")))
```

```python
nums
```

```python
def rewrite_numbers(nums):
    res = []
    n = len(nums)
    m = len(nums[0])
    for col in range(m):
        cur_num = ""
        for row in range(n):
            cur_num += nums[row][col]
        res.append(cur_num)
    return res
```

```python
def seperate_nums(nums):
    res = []
    temp = []
    for num in nums:
        if num:
            temp.append(int(num[0]))
        else:
            res.append(temp)
            temp = []
    res.append(temp)
    return res
```

```python
new_nums = rewrite_numbers(nums)
new_nums = [list(filter(None, x.split(" "))) for x in new_nums]
new_nums = seperate_nums(new_nums)
new_nums
```

```python
def prod_or_sum(array, operator):
    if operator == "*":
        return int(np.prod(array))
    elif operator == "+":
        return int(np.sum(array))
```

```python
prod_or_sum(new_nums[3], operators[3])
```

```python
res = 0
for array, operator in zip(new_nums, operators):
    res += prod_or_sum(array, operator)
res
```
