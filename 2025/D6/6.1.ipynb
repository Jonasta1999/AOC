```python
import numpy as np
```

```python
with open('input6.txt') as f:
    data = f.read().splitlines()
    data = [x.split(" ") for x in data]
    data = [list(filter(None, x)) for x in data]
    nums = [list(map(int, x)) for x in data[:-1]]
    nums = np.matrix(nums).transpose()
    operators = data[-1]
```

```python
def prod_or_sum(array, operator):
    if operator == "*":
        return int(np.prod(array))
    elif operator == "+":
        return int(np.sum(array))
```

```python
res = 0
for array, operator in zip(nums, operators):
    res += prod_or_sum(array, operator)
res
```
