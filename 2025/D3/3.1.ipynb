```python
with open('input3.txt') as f:
    data = f.read().splitlines()
    data = [list(map(int, x)) for x in data]
```

```python
def find_joltage(seq):
    max_val_index = seq.index(max(seq[:-1]))
    num_1 = seq.pop(max_val_index)
    num_2 = max(seq[max_val_index:])
    return int(str(num_1) + str(num_2))
```

```python
res = 0
for seq in data:
    res += find_joltage(seq)
res
```
