```python
with open('input3.txt') as f:
    data = f.read().splitlines()
    data = [list(map(int, x)) for x in data]
```

```python
def find_joltage(seq):
    res = ""
    max_val_index = 0
    for i in range(11, -1, -1):
        if i == 11:
            cut_seq = seq[max_val_index:-i]
        elif i == 0:
            max_val_index += 1
            cut_seq = seq[max_val_index:]
        else:
            max_val_index += 1
            cut_seq = seq[max_val_index:-i]
            
        new_max_val_index = cut_seq.index(max(cut_seq))
        max_val = cut_seq[new_max_val_index]
        max_val_index += new_max_val_index
        res += str(max_val)

    return int(res)
```

```python
res = 0
for seq in data:
    res += find_joltage(seq)
res
```
