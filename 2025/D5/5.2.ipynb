```python
with open('input5.txt') as f:
    data = f.read().splitlines()
    ind = data.index('')
    ranges = data[:ind]
    ranges = [list(map(int, range.split('-'))) for range in ranges]
    ingredients = list(map(int, data[ind+1:]))
```

```python
def merge_two_ranges(range1, range2):
    conditions = [
        range1[0] in range(range2[0], range2[1]+1),
        range1[1] in range(range2[0], range2[1]+1),
        range2[0] in range(range1[0], range1[1]+1),
        range2[1] in range(range1[0], range1[1]+1),
    ]
    if any(conditions):
        merged_range = (min(range1[0], range2[0]), max(range1[1], range2[1]))
        return merged_range
    else:
        return False
```

```python
different_ranges = True
merged_range = False
while different_ranges:
    for i in range(len(ranges)-1):
        for j in range(i+1, len(ranges)):
            merged_range = merge_two_ranges(ranges[i], ranges[j])
            if merged_range:
                ranges.pop(j)
                ranges.pop(i)
                ranges.append(merged_range)
                break
        if merged_range:
            break
    if merged_range:
        continue
    different_ranges = False
```

```python
count = [num2 - num1 + 1 for num1, num2 in ranges]
```
