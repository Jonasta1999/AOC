```python
with open('input5.txt') as f:
    data = f.read().splitlines()
    ind = data.index('')
    ranges = data[:ind]
    ranges = [list(map(int, range.split('-'))) for range in ranges]
    ingredients = list(map(int, data[ind+1:]))
```

```python
count = 0
for ingredient in ingredients:
    for r in ranges:
        if ingredient in range(r[0], r[1]+1):
            count+=1
            break

count
```
