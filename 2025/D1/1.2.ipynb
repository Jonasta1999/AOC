```python
with open('input1.txt') as f:
    data = f.read().splitlines()
```

```python
def valid_num(num, count):
    while num < 0 or num > 99:
        if num < 0:
            num += 100
            count += 1
        elif num > 99:
            num -= 100
            count += 1
    return num, count
```

```python
def find_password(data):
    password = 0
    n = len(data)
    pos = 50
    for i in range(n):
        direc = data[i][0]
        steps = int(data[i][1:])
        if direc == 'R':
            pos += steps
        elif direc == 'L':
            if pos == 0:
                pos = 100
            pos -= steps
        pos, password = valid_num(pos, password)
        if pos == 0 and direc == 'L':
            password += 1
    return password
```

```python
find_password(data)
```
