```python
with open('input1.txt') as f:
    data = f.read().splitlines()
```

```python
def valid_num(num):
    while num < 0 or num > 99:
        if num < 0:
            num += 100
        elif num > 99:
            num -= 100
    return num
```

```python
def find_password(data):
    password = 0
    n = len(data)
    pos = 50
    for i in range(n):
        direc = data[i][0]
        steps = int(data[i][1:])

        if direc == 'R':
            pos += steps
        elif direc == 'L':
            pos -= steps
        pos = valid_num(pos)

        if pos == 0:
            password += 1
    return password
```

```python
find_password(data)
```
