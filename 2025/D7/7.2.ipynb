```python
import numpy as np
```

```python
with open('input7.txt') as f:
    original_data = f.read().splitlines()
    data = [x.replace(".", "0") for x in original_data]
    data = [list(x) for x in data]
    data = np.array(data, dtype='<U10000')
    data = np.matrix(data)
```

```python
def beam_splitter(curr_pos):
    global data
    x, y = curr_pos
    if data[x-1, y] != "^" and data[x, y] != "^":
        data[x, y] = int(data[x, y]) + int(data[x-1, y])
    elif data[x-1, y] != "^" and data[x, y] == "^":
        data[x, y-1] = int(data[x, y-1]) + int(data[x-1, y])
        data[x, y+1] = int(data[x, y+1]) + int(data[x-1, y])
```

```python
start = (0, original_data[0].find("S"))
data[start] = "1"
```

```python
n, m = data.shape
for i in range(1, n):
    for j in range(m):
        beam_splitter((i,j))
```
