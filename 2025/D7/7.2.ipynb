```python
import numpy as np
```

```python
with open('input7.txt') as f:
    original_data = f.read().splitlines()
    data = [x.replace(".", "0") for x in original_data]
    data = [list(x) for x in data]
    data = np.array(data, dtype='<U10')
    data = np.matrix(data)
```

```python
print(data)
```

```python
count = 0
def beam_splitter(curr_pos):
    global count
    x, y = curr_pos
    try:
        if data[x+1, y] != "^":
            beam_splitter((x+1, y))
        elif data[x+1, y] == "^":
            beam_splitter((x, y-1))
            beam_splitter((x, y+1))
    except:
        count += 1
        pass
```

```python
start = (0, original_data[0].find("S"))
beam_splitter(start)
```
