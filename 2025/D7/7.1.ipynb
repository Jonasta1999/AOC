```python
with open('input7.txt') as f:
    data = f.read().splitlines()
```

```python
count = 0
def beam_splitter(data, curr_pos):
    x, y = curr_pos
    try:
        if data[x+1][y] == ".":
            data[x+1] = data[x+1][:y] + "|" + data[x+1][y+1:]
            beam_splitter(data, (x+1, y))
        elif data[x+1][y] == "^":
            global count
            count += 1
            beam_splitter(data, (x, y-1))
            beam_splitter(data, (x, y+1))
    except:
        pass
```

```python
start = (0, data[0].find("S"))
beam_splitter(data, start)
count
```
