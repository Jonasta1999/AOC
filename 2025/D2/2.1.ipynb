```python
with open('input2.txt') as f:
    data = f.read().split(",")
    data = [x.split("-") for x in data]
```

```python
def find_invalid_IDs(id1, id2):
    res = []
    for i in range(id1,id2+1):
        num_str = str(i)
        n = len(num_str)
        if n%2==0:
            ind = int(n/2)
            if num_str[:ind] == num_str[ind:]:
                res.append(i)
    return res
```

```python
res = []
n = len(data)
for i in range(n):
    id1 = int(data[i][0])
    id2 = int(data[i][1])
    res += find_invalid_IDs(id1, id2)
sum(res)
```
