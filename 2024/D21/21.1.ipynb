```python
num_pad = ["#####", "#789#", "#456#", "#123#", "##0A#", "#####"]
num_pad = np.asmatrix([list(item) for item in num_pad])
key_pad = ["#####", "##^A#", "#<v>#", "#####"]
key_pad = np.asmatrix([list(item) for item in key_pad])
num_pad
```

```python
def find_start_end(maze, start, end):
    n, m = maze.shape

    for i in range(n):
        for j in range(m):
            if maze[i, j] == start:
                start = (i, j)
            elif maze[i, j] == end:
                end = (i, j)
    return start, end

```

```python
direc_dict = {"^": (-1, 0),
              ">": (0, 1),
              "v": (1, 0),
              "<": (0, -1)}
```

```python
def get(node, maze, prev_direc):
    res_cord = []
    res_direc = []
    direcs = [((-1, 0), "^"), ((0, 1), ">"), ((1, 0), "v"), ((0, -1), "<")]
    #if prev_direc:
    #    prev_direc = prev_direc[-1]
    #    ind_pop_to_find = (direc_dict[prev_direc], prev_direc)
    #    ind_pop = direcs.index(ind_pop_to_find)
    #    first_direc = direcs.pop(ind_pop)
    #    direcs = [first_direc] + direcs
    for i in range(4):
        new_coord = (node[0]+direcs[i][0][0], node[1]+direcs[i][0][1])
        if maze[new_coord] != "#":
            res_cord.append(new_coord)
            res_direc.append(direcs[i][1])
    return res_cord, res_direc
```

```python
def bfs(start, end, maze):
    q = []
    q_direc = [[]]
    q.append([start])
    

    while q:
        current_path = q.pop(0)
        current_path_direc = q_direc.pop(0)
        node = current_path[-1]

        neighbours, direc = get(node, maze, current_path_direc)
        for i, next in enumerate(neighbours):
            if next not in current_path:
                temp_path = current_path.copy()
                temp_path.append(next)
                temp_path_direc = current_path_direc.copy()
                temp_path_direc.append(direc[i])

                if next == end:
                    return temp_path, temp_path_direc
                else:
                    q.append(temp_path)
                    q_direc.append(temp_path_direc)
    return [], []
    
```

```python
def flatten_nested_list(nested_list):
    result = []
    for element in nested_list:
        if isinstance(element, list):
            result.extend(flatten_nested_list(element))
        else:
            result.append(element)
    return result
```
